```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
crashes = pd.read_csv('./data/Traffic_Crashes_-_Crashes_20241213.csv', low_memory = False)
crashes.head()
```

```python
crashes.info()
```

```python
crashes.columns = crashes.columns.str.lower()
```

```python
# Convert all string values in object columns to lowercase
for col in crashes.select_dtypes(include='object').columns:
    crashes[col] = crashes[col].str.lower()
```

```python
for feature in crashes.columns:
    print(f"Value counts for column '{feature}':")
    print(crashes[feature].value_counts())
    
    # Print the number of unique values in the feature
    num_unique = crashes[feature].nunique()
    print(f"Number of unique values: {num_unique}")
    
    print("\n")
```

```python
missing_percentage = round((crashes.isna().sum()/len(crashes)*100), 2)
missing_percentage
```

```python
columns_with_missing_data = missing_percentage[missing_percentage > 50]
list(columns_with_missing_data.index)
```

```python
for feature in columns_with_missing_data.index:
    print(f"Value counts for column '{feature}':")
    print(crashes[feature].value_counts())
    print("\n")
```

```python
# Calculate the percentage of missing values for each column
missing_percentage = (crashes.isna().sum() / len(crashes)) * 100

# Filter columns with more than 80% missing values
columns_to_drop = missing_percentage[missing_percentage > 80].index

# Drop these columns from the dataframe
crashes_cleaned = crashes.drop(columns=columns_to_drop)

# Display the cleaned dataframe (optional)
crashes_cleaned.info()
```

```python
features_to_remove = ['date_police_notified', 
                     'crash_type',
                     'street_no',
                     'injuries_unknown',
                     'roadway_surface_cond',
                     'weather_condition']
```

```python
crashes_cleaned = crashes.drop(columns=features_to_remove)
```

```python
crashes_cleaned.info()
```

```python
# Convert the 'crash_date' column to datetime if it's not already in that format
crashes['crash_date'] = pd.to_datetime(crashes['crash_date'], errors='coerce')

# Get the earliest and latest entry in the 'crash_date' feature
earliest_date = crashes['crash_date'].min()
latest_date = crashes['crash_date'].max()

# Print the results
print(f"Earliest crash date: {earliest_date}")
print(f"Latest crash date: {latest_date}")
```

```python
# Group by longitude, latitude, and crash_date and count occurrences
duplicates = crashes.groupby(['longitude', 'latitude', 'crash_date']).size().reset_index(name='count')

# Filter for entries where the count is greater than 1 (duplicates)
duplicates = duplicates[duplicates['count'] > 1]

# Display the duplicates
duplicates
```
